## Population

```python
population = pd.read_csv(r"C:\Users\Harsh mishra\Desktop\solarFarm\data\population\Population of India.csv")

population.head()
```

```python
population.rename(columns = {'State/UT' : 'State'}, inplace = True)
population.rename(columns = {'Population[50]' : 'Population'}, inplace = True)
population.rename(columns = {'Rural[51]' : 'Rural'}, inplace = True)
population.rename(columns = {'Urban[51]' : 'Urban'}, inplace = True)
population.rename(columns = {'Area[52] (km2)' : 'Area'}, inplace = True)

print("Before dropping")
print(population.columns)

population.drop(population.columns[[3, 4, 5, 6, 7]], axis = 1, inplace = True)

print("After dropping")
print(population.columns)
```

## Solar

```python
solar = pd.read_csv(r"C:\Users\Harsh mishra\Desktop\solarFarm\data\solar_farms_india_2021.csv")

solar.head()
```

Common column

```python
import numpy as np

common = np.intersect1d(solar.columns, population.columns)

print(common)
```

```python
merged_df_outer = pd.merge(population, solar, on='State', how='outer')

print(merged_df_outer)
```

```python
null_count = merged_df_outer.isnull().sum().sum()
print('Number of null values:', null_count)

merged_df_outer["Total Solar Farm Count"].fillna(0, inplace = True) 
merged_df_outer["Total Solar Area (km2)"].fillna(0, inplace = True) 
print('Number of null values:', null_count)
```

```python
print(merged_df_outer)
```

```python
print(merged_df_outer.columns)
```

```python


merged_df_outer['Percentage'] = merged_df_outer['Total Solar Farm Count'] / merged_df_outer['Population']

print(merged_df_outer)
```
