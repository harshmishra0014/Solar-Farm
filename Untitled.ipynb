## Display Tiff file of LandCover

Here I am loading the tiff file, this tiff file is basically the landcover file of Indian region which gives information about geography.

```python
import imageio.v2 as imageio
import matplotlib.pyplot as plt


# Read the TIFF file
image = imageio.imread(file_path_landCover)

# Display the image using matplotlib
plt.imshow(image)
plt.axis('off')  # Hide axes
plt.show()
```

```python
import rasterio
import matplotlib.pyplot as plt

# Open the TIFF file using rasterio
with rasterio.open(file_path_landCover) as src:
    # Read all bands
    bands = src.read()

# Display each band separately
num_bands = bands.shape[0]

for band_num in range(1, num_bands + 1):
    plt.figure(figsize=(8, 8))
    plt.imshow(bands[band_num - 1], cmap='viridis')  # You can choose a different colormap
    plt.title(f'Band {band_num}')
    plt.colorbar()
    plt.show()
```

## GeoJson

The objective in this section is to open the geojson file visualise it.

```python
import geopandas as gpd
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
```

```python
# Load GeoJSON data
solar_farms = gpd.read_file(file_path_Geo)

# Load land cover TIFF file
with rasterio.open(file_path_landCover) as src:
    landcover = src.read(1)  # Assuming a single-band TIFF file

# Overlay solar farm locations on land cover
fig, ax = plt.subplots(figsize=(10, 10))
show(landcover, ax=ax)
solar_farms.plot(ax=ax, color='red', marker='o', markersize=50)
plt.show()
```
