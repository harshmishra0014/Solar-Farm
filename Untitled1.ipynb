## Distance between two points

```python
import math

def haversine(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dLat = lat2 - lat1
    dLon = lon2 - lon1
    a = math.sin(dLat / 2) ** 2 + math.cos(lat1) * math.cos(lat2) * math.sin(dLon / 2) ** 2
    c = 2 * math.asin(math.sqrt(a))
    radius_of_earth = 6371  # Earth's radius in kilometers

    return radius_of_earth * c

# Example usage
lat1, lon1 = 26.42317445549239, 80.3233615453453  # Coordinates for Big Ben in London
lat2, lon2 = 26.448934285933866, 80.19325398783036  # Coordinates for The Statue of Liberty in New York
distance_km = haversine(lat1, lon1, lat2, lon2)

print(f"Distance: {distance_km:.2f} km")
```

## All points at a certain radius

```python
import numpy as np

def generate_coordinates_within_radius(lat, lon, radius_km):
    # Constants
    earth_radius_km = 6371.0  # Approximate Earth radius in kilometers

    # Calculate the angular distance in radians
    radius_rad = radius_km / earth_radius_km

    # Convert radius from radians to degrees
    radius_deg = radius_rad * (180.0 / np.pi)

    # Create arrays of latitudes and longitudes within the specified radius
    lat_range = np.arange(lat - radius_deg, lat + radius_deg, 0.001)
    lon_range = np.arange(lon - radius_deg / np.cos(np.radians(lat)), lon + radius_deg / np.cos(np.radians(lat)), 0.001)

    coordinates = []
    for lat_point in lat_range:
        for lon_point in lon_range:
            # Calculate the distance between the center point and the generated point
            distance = haversine_distance(lat, lon, lat_point, lon_point)
            if distance <= radius_km:
                coordinates.append((lat_point, lon_point))
    
    return coordinates

def haversine_distance(lat1, lon1, lat2, lon2):
    # Haversine formula to calculate the distance between two points on the Earth
    R = 6371.0  # Earth radius in kilometers

    dlat = np.radians(lat2 - lat1)
    dlon = np.radians(lon2 - lon1)
    a = np.sin(dlat / 2)**2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c
    
    return distance

# Example usage
latitude, longitude = 26.42317445549239, 80.3233615453453
radius = 16  # 10 kilometer

coordinates_within_radius = generate_coordinates_within_radius(latitude, longitude, radius)

print(coordinates_within_radius)
```
